2. Limpieza de los datos:
- Eliminar datos no numéricos.
- Eliminar filas incompletas.

```python
ruta_de_archivos = '/content/drive/MyDrive/UPSO/MLyBD/Dataset de ventas'
archivos = os.listdir(ruta_de_archivos)

dataframes = {}
df_anual=[]
for archivo in archivos:
    if archivo.endswith('.csv'):  # Asegurarse de que solo se consideren archivos CSV
        mes = archivo.split('_')[3].split('.')[0]  # Obtener el nombre del archivo sin la extensión
        ruta_completa = os.path.join(ruta_de_archivos, archivo)  # Ruta completa al archivo
        df = pd.read_csv(ruta_completa)  # Cargar el archivo CSV en un DataFrame
        df_ventas = df.copy()

        df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
        df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')


        df_ventas = df_ventas.loc[~df_ventas['Cantidad Pedida'].isna()]
        df_ventas = df_ventas.loc[~df_ventas['Precio Unitario'].isna()]

        # Elimino todos los Nan y me quedo solo con los valores numericos
        df_ventas = df_ventas.dropna()

        # Transformo todos los valores numericos al tipo de dato int
        df_ventas['Cantidad Pedida'] = df_ventas['Cantidad Pedida'].astype(int)
        df_ventas['Precio Unitario'] = df_ventas['Precio Unitario'].astype(int)

        dataframes[mes] = df_ventas  # Agrego el DataFrame al diccionario usando el nombre del mes
        df_anual.append(dataframes[mes])
```

```python
df_anual = pd.concat(df_anual, ignore_index=True)
```

```python
#Extraer los datos de la columna 'Fecha de Pedido'

def extraer_datosFP(datos):
  datos['Fecha de Pedido'] = pd.to_datetime(datos['Fecha de Pedido'])
  datos['Hora'] = datos['Fecha de Pedido'].dt.hour
  datos['Dia'] = datos['Fecha de Pedido'].dt.day
  datos['Mes'] = datos['Fecha de Pedido'].dt.month
  datos['Año'] = datos['Fecha de Pedido'].dt.year
  columnas_a_eliminar = ['Fecha de Pedido']
  datos = datos.drop(columnas_a_eliminar, axis=1)

  return datos

#Extraigo los datos de la columna 'Fecha de Pedido' del dataframe anual
dataframe_anualFP = extraer_datosFP(df_anual)

#Extraigo los datos de la columna 'Fecha de Pedido' del dataframe mensual
dataframe_mesFP =extraer_datosFP(dataframe_mes)
```

```python
# Separar la columna 'Direccion de envio' en columnas de dirección, estado y código postal.

def extraer_datosDDE(datos):
  direccion_df = datos['Dirección de Envio'].str.split(',', expand=True)
  direccion_df.columns = ['Direccion', 'Ciudad', 'CodigoPostal']

  #Elimino los espacios al comienzo de cada celda de la columna CodigoPostal.
  direccion_df['CodigoPostal'] = direccion_df['CodigoPostal'].str.strip()

  #Separo la columna 'CodigoPostal' en Estado y CodigoPostal.
  codigopostal_df = direccion_df['CodigoPostal'].str.split(' ',expand=True)
  codigopostal_df.columns = ['Estado', 'CodPostal']
  datos = pd.concat([datos, direccion_df, codigopostal_df], axis=True)
  columnas_a_eliminar = ['Dirección de Envio', 'CodigoPostal']
  datos = datos.drop(columnas_a_eliminar, axis=1)
  return datos

#Extraigo los datos de la columna 'Fecha de Pedido' del dataframe anual
dataframe_anualDDE = extraer_datosDDE(dataframe_anualFP)

#Extraigo los datos de la columna 'Fecha de Pedido' del dataframe mensual
dataframe_mesDDE = extraer_datosDDE(dataframe_mesFP)
```

2. Optimización de la publicidad y patrón de ventas por hora:

  ● ¿Cuál es el momento ideal del día para presentar la publicidad y aumentar la
  probabilidad de compra?

  ● ¿Cómo cambian los patrones de ventas por hora a lo largo del año?

  ● ¿Hay modificaciones en los patrones de ventas durante las horas de mayor actividad en los distintos meses?

```python
# Creo una lista para almacenar los resultados por mes
resultados_por_mes = []

# Creo una variable para almacenar los resultados de todo el año
resultados_totales = None

for mes in meses:
    dataframe_mes = dataframes[mes]
    dataframe_mes = extraer_datosFP(dataframe_mes)

    # Agrupo los datos por hora y suma la cantidad pedida de productos
    ventas_por_hora = dataframe_mes.groupby('Hora')['Cantidad Pedida'].sum()

    # Encuentro la hora en la que se vendió más en este mes
    hora_maxima_mes = ventas_por_hora.idxmax()
    cantidad_maxima_mes = ventas_por_hora.max()

    # Agrego los resultados por mes a la lista
    resultados_por_mes.append({
        'Mes': mes,
        'Hora_Mas_Ventas': hora_maxima_mes,
        'Cantidad_Mas_Ventas_Mes': cantidad_maxima_mes
    })

    # Agrega los datos de este mes al resultado total
    if resultados_totales is None:
        resultados_totales = ventas_por_hora
    else:
        resultados_totales += ventas_por_hora

# Encuentro la hora en la que se vendió más en todo el año
hora_maxima_anual = resultados_totales.idxmax()
cantidad_maxima_anual = resultados_totales.max()

# Agrego los resultados totales al resultado final
resultados_por_mes.append({
    'Mes': 'Anual',
    'Hora_Mas_Ventas': hora_maxima_anual,
    'Cantidad_Mas_Ventas_Mes': cantidad_maxima_anual
})

# Convierto los resultados a un DataFrame
resultado_final = pd.DataFrame(resultados_por_mes)

#Creo un gráfico de barras
fig = px.bar(resultado_final, x='Mes', y='Cantidad_Mas_Ventas_Mes', text='Hora_Mas_Ventas',
             title='Cantidad Máxima de Ventas por Mes',
             color='Mes',  # Colorear por la columna 'Mes'
             color_discrete_map={'Anual': 'red'}  # Color de la barra anual
             )

# Personalizo el diseño
fig.update_traces(textposition='outside')  # Mostrar la hora en el gráfico
fig.update_xaxes(type='category')  # Establecer el eje X como categorías (meses)
fig.update_layout(xaxis_title='Mes', yaxis_title='Cantidad Máxima de Ventas', title='CANTIDAD MAXIMA DE VENTAS POR MES Y HORA DE MAS VENTA')

# Muestro el gráfico de barras
fig.show()
```

```python
import plotly.express as px

# Convierto las columnas 'dia', 'mes' y 'año' en una columna de fecha completa y renombro para poder usar dt.dayofweek
dataframe_anualFP['Fecha'] = pd.to_datetime(dataframe_anualFP[['Dia', 'Mes', 'Año']].rename(columns={'Dia': 'Day','Mes':'Month', 'Año':'Year'}))

# Agrego una columna para el día de la semana (0: Lunes, 1: Martes, ..., 6: Domingo).
dataframe_anualFP['dia_semana'] = dataframe_anualFP['Fecha'].dt.dayofweek

# Agrupo los datos por día de la semana y suma la cantidad pedida.
ventas_total_por_dia_semana = dataframe_anualFP.groupby('dia_semana')['Cantidad Pedida'].sum()

# Nombres de los días de la semana para la visualización.
nombres_dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Crear un gráfico de líneas para visualizar las ventas por día de la semana.
fig_ventas_dias = go.Figure(data=go.Scatter(x=nombres_dias, y=ventas_total_por_dia_semana, mode='lines'))
fig_ventas_dias.update_layout(
    title='VENTAS ANUALES POR DIA DE LA SEMANA',
    xaxis_title='Día de la Semana',
    yaxis_title='Ventas',
)

# Muestro el gráfico de líneas
fig_ventas_dias.show()
```

```python
# Ventas mensual por dia de la semana

# Creo una lista para almacenar los datos de los gráficos de histograma
graficos_histograma = []

# Itero a través de los meses
for mes in meses:
    dataframe_mes = dataframes[mes]
    dataframe_mes = extraer_datosFP(dataframe_mes)
    dataframe_mes['Fecha'] = pd.to_datetime(dataframe_mes[['Dia', 'Mes', 'Año']].rename(columns={'Dia': 'Day', 'Mes': 'Month', 'Año': 'Year'}))
    dataframe_mes['dia_semana'] = dataframe_mes['Fecha'].dt.dayofweek

    # Obtengo las ventas por día de la semana
    ventas_por_dia_semana = dataframe_mes.groupby('dia_semana')['Cantidad Pedida'].sum()

    # Creo un gráfico de histograma de barras
    grafico_histograma = go.Bar(x=nombres_dias, y=ventas_por_dia_semana, name=mes)

    # Agrego el gráfico a la lista
    graficos_histograma.append(grafico_histograma)

# Creo la figura con todos los gráficos de histograma
fig_ventas_dias_histograma = go.Figure(data=graficos_histograma)
fig_ventas_dias_histograma.update_layout(
    barmode='group',  # Agrupar las barras
    title='VENTAS MENSUALES POR DIAS DE LA SEMANA',
    xaxis_title='Día de la Semana',
    yaxis_title='Ventas',
)

# Muestro la figura con los gráficos de histograma de barras de distintos colores
fig_ventas_dias_histograma.show()
```
